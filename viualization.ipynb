```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
```

```python
%pip install wordcloud
from wordcloud import WordCloud
```

```python
data = pd.read_excel('excele_analytic\privitive_rubrics_frequency_sorted.xlsx')
```

```python
words = data['split_rubrics']
count = data['count']

dic = dict(zip(words, count))

wordcloud_word = WordCloud(background_color='black').generate_from_frequencies(dic)

plt.figure()
plt.figure(figsize=(5,5))
plt.imshow(wordcloud_word)
plt.axis("off")
plt.show()
```
