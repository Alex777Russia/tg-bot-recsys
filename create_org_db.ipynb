```python
db = pd.read_csv('d:\\BMSTU_im_Baumana\\семестры\\Семестра_5\\nlp\\with_parser.csv', sep='\t')
db = db.drop('Unnamed: 0', axis=1)
db
```

```python
db[db['name_ru'] == 'Воздух']
```

```python
db_gb = db.groupby(["adress", "name_ru", "rubrics"])
db_gb
```

```python
db_gb_mean = db_gb['rating'].mean()
db_gb_mean
```

```python
type(db_gb_mean)
```

```python
db_gb_count = db_gb.count()
db_gb_count
```

```python
db_gb_count[db_gb_count['rating'] == 1]
```

```python
print(f"Максимальное количество отзывов по одной из организаций: {db_gb_count.rating.max()}")
```

```python
import matplotlib.pyplot as plt

db_gb_count.sort_values(by=['rating'], inplace=True)
db_gb_count
```

```python
import seaborn as sns

sns.countplot(x='rating', data=db_gb_count)

sns.set(rc={'figure.figsize':(27.7, 8.27)})

plt.title("Frequency of values in 'rating'")
plt.xlabel("Value of reviews")
plt.ylabel("Count of organizations")
plt.show()
```

```python
sns.countplot(x='rating', data=db_gb_count)
sns.set(rc={'figure.figsize':(11.7,8.27)})

plt.title("Frequency of values in 'rating'")
plt.xlabel("Value of reviews")
plt.ylabel("Count of organizations")
plt.show()
```

```python
concat = pd.concat([db_gb_count, db_gb_mean], axis=1).reset_index()
concat
```

```python
concat.columns = ['adress', 'name_ru', 'rubrics', 'count_reviews', 'text', 'mean_rating']
concat = concat.drop('text', axis=1)
concat
```

```python
concat.to_csv('organizations.csv')
```
