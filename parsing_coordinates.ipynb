```python
from dadata import Dadata
import pandas as pd
```

```python
db = pd.read_csv('moscow_unique_adress.csv')
first = db['adress'].tolist()
first
```

```python
print(f"Проверка на длину списка адрессов: {len(first)}")
```

```python
needed_arr = first[16788:25000]
needed_arr
```

```python
l = len(needed_arr)
print(f"Длина списка, который мы подадим в api: {l}")
```

```python
# samoware shad21...
token = "481d7e0beb7d213d3358c5c20d6ab6fa5faeb8be"
secret = "e5a68440fedb760018492b5802ddcb88c1c743fe"

# # samoware arsen
# token = 
# secret = 

dadata = Dadata(token, secret)

result_adress = []

for i in range(l):
    result = dadata.clean("address", needed_arr[i])
    if 'geo_lat' in result:
        result_adress.append(result['geo_lat'] + " " + result['geo_lon'])
    else:
        print(f"Index {i} не имеет адресса")
```

```python
res_l = len(result_adress)
print(f"Длина списка, который мы получили от api: {res_l}")
```

Сохранение получившихся результатов в последовательном порядке в файл

```python
with open('file_16788_21814.txt', 'w') as f:
    for line in result_adress:
        f.write(f"{line}\n")
```
