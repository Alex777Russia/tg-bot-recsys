```python
from dadata import Dadata
import pandas as pd
```

```python
db = pd.read_csv('D:\\BMSTU_im_Baumana\\семестры\\Семестра_5\\nlp\\bert-01\\moscow_unique_adress.csv')
first = db['adress'].tolist()
first
```

```python
print(f"Проверка на длину списка адрессов: {len(first)}")
```

```python
needed_arr = first[11041:12000]
needed_arr
```

```python
l = len(needed_arr)
print(f"Длина списка, который мы подадим в api: {l}")
```

```python
token = "e54ac2506805e632bb565d9dae5ffe33ec1345f5"
secret = "2094053963d6410fecaca8e72dc18ea8e3c40d18"
dadata = Dadata(token, secret)

result_adress = []

for i in range(l):
    result = dadata.clean("address", first[i])
    result_adress.append(result['geo_lat'] + " " + result['geo_lon'])
```

```python
res_l = len(result_adress)
print(f"Длина списка, который мы получили от api: {res_l}")
```
